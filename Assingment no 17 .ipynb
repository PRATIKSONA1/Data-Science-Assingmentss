```python
# 1. Assign the value 7 to the variable guess_me. Then, write the conditional tests (if, else, and elif) to
# print the string &#39;too low&#39; if guess_me is less than 7, &#39;too high&#39; if greater than 7, and &#39;just right&#39; if equal
# to 7.
Ans=
Here's an example of how you can write the conditional tests to print the appropriate string based on the value of guess_me:


guess_me = 7

if guess_me < 7:
    print('too low')
elif guess_me > 7:
    print('too high')
else:
    print('just right')
In this code, the first if statement checks if guess_me is less than 7. If it is, it prints 'too low'.
The elif statement checks if guess_me is greater than 7. If it is, it prints 'too high'. If neither of 
these conditions is true, the else statement is executed and it prints 'just right'.
```

```python
# 2. Assign the value 7 to the variable guess_me and the value 1 to the variable start. Write a while
# loop that compares start with guess_me. Print too low if start is less than guess me. If start equals
# guess_me, print &#39;found it!&#39; and exit the loop. If start is greater than guess_me, print &#39;oops&#39; and exit
# the loop. Increment start at the end of the loop.

Ans=
Here's an example of how you can write a while loop to compare start with guess_me and print the corresponding messages:


guess_me = 7
start = 1

while start <= guess_me:
    if start < guess_me:
        print('too low')
    elif start == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    start += 1
In this code, the while loop continues as long as start is less than or equal 
to guess_me. Inside the loop, it checks the conditions using if, elif, and else statements.
If start is less than guess_me, it prints 'too low'. If start is equal to guess_me, it prints
'found it!' and exits the loop using the break statement. If start is greater than guess_me, 
it prints 'oops' and exits the loop. At the end of each iteration, the value of start is 
incremented by 1 using start += 1.
```

```python
# 4. Use a list comprehension to make a list of the even numbers in range(10)
Ans=
You can use a list comprehension to create a list of even numbers in the range from 0 to 9 (i.e., range(10)). Here's an example:


even_numbers = [num for num in range(10) if num % 2 == 0]
print(even_numbers)
Output:


[0, 2, 4, 6, 8]
```

```python
# 5. Use a dictionary comprehension to create the dictionary squares. Use range(10) to return the
# keys, and use the square of each key as its value.
Ans=
You can use a dictionary comprehension to create a dictionary called squares where the keys are the numbers from 0 to 9 (using range(10)) and the values are the squares of the corresponding keys. Here's an example:


squares = {num: num**2 for num in range(10)}
print(squares)
Output:


{0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64, 9: 81}
```

```python
# 6. Construct the set odd from the odd numbers in the range using a set comprehension (10).
Ans=

You can use a set comprehension to create a set called odd that contains the odd numbers from the range 0 to 9 (using range(10)). Here's an example:


odd = {num for num in range(10) if num % 2 != 0}
print(odd)
Output:


{1, 3, 5, 7, 9}
```

```python
# 7. Use a generator comprehension to return the string &#39;Got &#39; and a number for the numbers in
# range(10). Iterate through this by using a for loop.
Ans=
You can use a generator comprehension to generate a sequence of strings combining the string "Got " with each number in the range 0 to 9 (using range(10)). Here's an example:


generator = ('Got ' + str(num) for num in range(10))

for item in generator:
    print(item)
Output:


Got 0
Got 1
Got 2
Got 3
Got 4
Got 5
Got 6
Got 7
Got 8
Got 9
```

```python
# 9. Define a generator function called get_odds that returns the odd numbers from range(10). Use a
# for loop to find and print the third value returned.
Ans=
Here's an example of how you can define a generator function called get_odds that yields the odd numbers from the range 0 to 9 (using range(10)). Then, using a for loop, you can find and print the third value returned by the generator:


def get_odds():
    for num in range(10):
        if num % 2 != 0:
            yield num

# Using a for loop to find and print the third value returned by the generator
count = 0
for odd_num in get_odds():
    count += 1
    if count == 3:
        print(odd_num)
        break

5
In this code, the get_odds function is defined as a generator function using the yield keyword.
It iterates over the numbers from 0 to 9 using range(10), and if a number is odd (determined by num % 2 != 0), 
it yields the number. The for loop is then used to iterate over the generator. In each iteration, the count variable is incremented, and when count reaches 3, the third
value returned by the generator is printed using the print() function, and the loop is terminated using the break statement.
```

```python
# 10. Define an exception called OopsException. Raise this exception to see what happens. Then write
# the code to catch this exception and print &#39;Caught an oops&#39;.
Ans=
Here's an example of how you can define an exception called OopsException, raise it to see what happens, and then catch it using a try-except block to print the message 'Caught an oops':


class OopsException(Exception):
    pass

try:
    raise OopsException
except OopsException:
    print('Caught an oops')
Output:


In this code, the OopsException class is defined by inheriting from the Exception class. This creates a custom exception called OopsException.

Inside the try block, the raise statement is used to raise an instance of the OopsException. This causes the exception to be raised.

In the except block, we catch the OopsException using the except clause with the exception class name. When the OopsException is caught, the code inside the except
block is executed, which in this case simply prints the message 'Caught an oops'.
```

```python
# 11. Use zip() to make a dictionary called movies that pairs these lists: titles = [&#39;Creature of Habit&#39;,
# &#39;Crewel Fate&#39;] and plots = [&#39;A nun turns into a monster&#39;, &#39;A haunted yarn shop&#39;].
Ans=
You can use the zip() function along with the dict() constructor to create a dictionary called movies by pairing the elements from the titles and plots lists. Here's an example:


titles = ['Creature of Habit', 'Crewel Fate']
plots = ['A nun turns into a monster', 'A haunted yarn shop']

movies = dict(zip(titles, plots))
print(movies)
Output:


{'Creature of Habit': 'A nun turns into a monster', 'Crewel Fate': 'A haunted yarn shop'}
In this code, the zip() function is used to pair corresponding elements from the titles and plots lists together. The resulting iterable of tuples is then passed to the dict() constructor, which creates a dictionary where the elements from titles are the keys and the elements from plots are the values. The resulting dictionary movies contains the pairs of titles and plots.






```
